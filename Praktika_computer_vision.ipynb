```python
import cv2
from matplotlib import pyplot as plt

#load an image from file
image = cv2.imread('cat.jpg')

#Display the image
plt.imshow(image)
```

```python
#Load an image from file
image = cv2.imread('cat.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Display the image
plt.imshow(image)
```

```python
print("Image demensions:", image.shape)
```

```python
# Resize the image to specific dimension

resized_image_specific = cv2.resize(image, (230,300))
plt.imshow(resized_image_specific)

print("Image demensions:", resized_image_specific.shape)
```

```python
# Resize image to 25%
resized_image_procent = cv2.resize(image, (0,0), fx=0.25, fy=0.25)
plt.imshow(resized_image_procent)

print("Image demensions:", resized_image_procent.shape)
```

```python
# Resize image to 400 * 400

resized_image = cv2.resize(image, (400,400))

plt.imshow(resized_image)

print("Image resized:", resized_image.shape)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
# Save the grayscale image to a file

cv2.imwrite('gray_cat.jpg', gray_image)
```

```python
# Draw a rectangle on the image (top-left corner at (50, 50))
# bottom-right corner (200, 200)
image = image.copy()
cv2.rectangle(image, (50, 50), (200, 200), (255, 0, 0), 5)
plt.imshow(image)
```

```python
cv2.circle (image, (600, 600), (200), (0, 0, 250), thickness=25)
plt.imshow(image)
```
